```python
import gradio as gr

def add_numbers(a, b):
    return a + b

demo = gr.Interface(
    fn=add_numbers,              # Hàm xử lý
    inputs=[gr.Number(), gr.Number()],  # Input: 2 số
    outputs=gr.Number(),         # Output: 1 số
    title="Simple Adder",
    description="Enter two numbers to add."
)

demo.launch(share=True, debug=True, inline=False)  # share=True để tạo link public
# Luôn dùng share=True để có link truy cập từ bên ngoài.
# Nếu quên share=True, giao diện chỉ hiển thị trong nội bộ Colab (khó dùng).
# Nếu bạn chạy nhiều lần, Gradio có thể báo port conflict → chỉ cần restart runtime hoặc thêm demo.launch(share=True, inline=False).

# Đọc thêm tài liệu ve
```

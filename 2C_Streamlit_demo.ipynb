```python
%%writefile app.py
import streamlit as st

st.title("Simple Streamlit Demo in Colab")

a = st.number_input("Enter first number", 0)
b = st.number_input("Enter second number", 0)

if st.button("Add"):
    st.success(f"Result: {a + b}")
```

```python
from pyngrok import ngrok

#TODO: Vào ngrok.com đăng kí một account để lấy API-KEY
from google.colab import userdata
ngrok_api_key = userdata.get('ngrok-api-key')
!ngrok config add-authtoken $ngrok_api_key

# Mở tunnel cho port 8501 (mặc định của Streamlit)
public_url = ngrok.connect(8501)
print("Streamlit URL:", public_url)
```
